# 3.) Clean your data for a decision tree


```python
features_cat = features[['class_of_admission','application_type','pw_level_9089','country_of_citizenship']]
features_num = pd.get_dummies(features_cat)
features_final = pd.concat([features_num, features['case_status']], axis=1)
```

```python
features_final
```

# 4.) Fit and plot a decision tree of depth 3


```python
X = features_final.drop('case_status', axis = 1).astype(int)
y = features_final[['case_status']]
```

```python
from sklearn import tree
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = tree.DecisionTreeClassifier(max_depth = 3)
clf = clf.fit(X,y)
```

# 5.) Write your interpretation of the largest (by sample size) leaf node


```python
plt.figure(figsize=(20,10))
tree.plot_tree(clf, rounded=True, feature_names=X.columns)
```

```python
value_counts = features_final['case_status'].value_counts()
print(value_counts)
```

The largest leaf node has 83522 samples and a gini value of 0.603, this node is determined by the feature 'country_of_citizenship_PHILIPPINES'. When applicants' country of citizenship is PHILIPPINES, the result is most likely "certified". However, the gini value of 0.603 indicates that the diversity of the target class values within this node is not pure and there is a mix of different case statuses.

# 6.) Using a for loop, make your own train-test split and determine the best max_depth for out-of sample accuracy

```python
from sklearn.metrics import confusion_matrix, f1_score
from sklearn.metrics import precision_recall_curve
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import metrics
```

```python
max_depths = [1,2,3]
outputs = []

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)
for md in max_depths:
  clf = tree.DecisionTreeClassifier(max_depth = md)
  clf.fit(X_train,y_train)

  y_pred = clf.predict(X_test)
  oos_accuracy = metrics.accuracy_score(y_test, y_pred)
  outputs.append(oos_accuracy)

best_max_depth = max_depths[outputs.index(max(outputs))]
print("Best max_depth: ", best_max_depth)
```
